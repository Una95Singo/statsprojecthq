```python
dfEventTypes = pd.read_csv("/Users/singo/Downloads/vulaactivity/VULA-3233_event_types.csv", index_col=0)
```

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import collections
```

```python
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
sns.countplot(data=df, x='EVENT_ID' ,  palette="Greens_d",   )
sns.set(font_scale=1, rc={'figure.figsize':(35,30),"xtick.major.size": 8})
```

```python
mostCommon = collections.Counter(df.EVENT_ID).most_common(50)
for item in mostCommon:
    print(str(dfEventTypes.loc[item[0],'DESCRIPTION']) + " " + str(item[1]))
```

```python
randomStudent = df.groupby("USER_ID").get_group('00269aa22374cd8666ce622800fb933ed1e152ff6e31fc05e9e43cc1335c5e99')

sns.countplot(data=randomStudent , x='EVENT_ID' ,  palette="Greens_d",   )
sns.set(font_scale=1, rc={'figure.figsize':(35,30),"xtick.major.size": 8})
```

```python
mostCommon = collections.Counter(randomStudent.EVENT_ID).most_common(50)
for item in mostCommon:
    print(str(dfEventTypes.loc[item[0],'DESCRIPTION']) + " " + str(item[1]))
```
